```python
!pip install mtcnn
from mtcnn.mtcnn import MTCNN
import cv2
from matplotlib import pyplot as plt
import matplotlib.patches as patches

# Load the image
image_path = '/content/0a1a955cb3527076381726e8c898c6f8.jpeg'
image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)

# Initialize the detector
detector = MTCNN()

# Detect faces
result = detector.detect_faces(image)

# Plot the image and the bounding boxes
fig, ax = plt.subplots(1)
ax.imshow(image)
for face in result:
    x, y, width, height = face['box']
    rect = patches.Rectangle((x, y), width, height, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
plt.axis('off')
plt.savefig('detected_faces.png')
plt.show()
```

```python
import cv2

# Load the image
image = cv2.imread('detected_faces.png')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load the Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Print the coordinates of the bounding box for each face
for (x, y, w, h) in faces:
    print(f"Bounding box coordinates: x={x}, y={y}, width={w}, height={h}")
```
