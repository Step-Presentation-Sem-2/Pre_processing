### Getting Real Scraped Images from S3 Bucket

```python
import boto3
import os

# Define the clean_file_name function to handle path and redundant extension issues
def clean_file_name(file_name):
    # Normalize the file path to ensure correct path separators
    file_name = os.path.normpath(file_name)
    base_file_name = os.path.basename(file_name)

    # Remove redundant extensions
    while base_file_name.count('.jpg') > 1:
        base_file_name = base_file_name.rsplit('.jpg', 1)[0]
    while base_file_name.count('.png') > 1:
        base_file_name = base_file_name.rsplit('.png', 1)[0]

    return base_file_name

def download_images_from_s3(bucket_name, local_folder):
    session = boto3.Session(
        aws_access_key_id=os.getenv("aws_access_key_id"),
        aws_secret_access_key=os.getenv("aws_secret_access_key")
    )
    s3 = session.client('s3')

    objects = s3.list_objects_v2(Bucket=bucket_name).get('Contents', [])
    if not objects:
        print("No objects found in the bucket.")
        return

    for obj in objects:
        file_name = obj['Key']
        clean_name = clean_file_name(file_name)
        local_file_path = os.path.join(local_folder, clean_name)
        
        # Ensure the directory for the file exists
        os.makedirs(os.path.dirname(local_file_path), exist_ok=True)
        
        try:
            if clean_name.endswith('.jpg') or clean_name.endswith('.png'):
                s3.download_file(bucket_name, file_name, local_file_path)
                print(f"{file_name} downloaded to {local_file_path}")
        except Exception as e:
            print(f"Failed to download {file_name}: {e}")

# Adjust the local_folder to use an absolute path
local_folder = 'C:/Users/ArshiaAchar/Desktop/DS_projects/Step2/Pre_processing/scraped_images_b1'
os.makedirs(local_folder, exist_ok=True)  # Ensure the directory is created

# Specify the correct bucket name
bucket_name = 'scraped-images-b1'

download_images_from_s3(bucket_name, local_folder)
```
